```python
import numpy as np
import pandas as pd
import time
import os
import cv2
import matplotlib.pyplot as plt
import cyvlfeat
from cyvlfeat import sift
```

```python
labels = ['non-COVID', 'COVID']
data_dir = '/Users/makhanov/Yandex.Disk.localized/NU-PhD/Fall2020/ComputerVision/project/datasets/archive'
```

```python
phase = 'train'
train_data = []

training_dir_path = (data_dir + '/' + phase)
training_names = os.listdir(training_dir_path)

train_image_paths = []
train_image_classes = []
train_class_id = 0
```

```python
def imglist(path):
    return [os.path.join(path, f) for f in os.listdir(path)]
```

```python
for training_name in training_names:
    dir = os.path.join(training_dir_path, training_name)
    if os.path.isdir(dir):
        class_path = imglist(dir)
        train_image_paths += class_path
        train_image_classes += [train_class_id]*len(class_path)
        train_class_id += 1
```

```python
des_list = []
```

```python
train_image_paths[144]
```

```python
counter = 0
for image_path in train_image_paths:
    im = cv2.imread(image_path, 0)
#     print(im.shape, counter)
    #kpts, des = brisk.detectAndCompute(im, None)
    kpts, des = sift.sift(im, compute_descriptor=True)
    des_list.append((image_path, des))
    counter+=1
```

```python
#Stack all the descriptors vertically in a numpy array
descriptors = des_list[0][1]
for image_path, descriptor in des_list[1:]:
    descriptors = np.vstack((descriptors, descriptor))
```

```python
#kmeans works only on float, so convert integers to float
descriptors_float = descriptors.astype(float)
```

```python
#k-means clustering and vector quatization
from scipy.cluster.vq import kmeans, vq
```

```python
k = 200
voc, variance = kmeans(descriptors_float, k, 1)
```

```python
#calculate the historgram of features and represent them as vector
#vq Assings codes from a code book to observations
im_features = np.zeros((len(train_image_paths), k), "float32")
for i in range(len(train_image_paths)):
    words, distance = vq(des_list[i][1], voc)
    for w in words:
        im_features[i][w] += 1
```

```python
im_features
```

```python
#perform Tf-Idf vectorization
nbr_occurences = np.sum((im_features > 0) * 1, axis=0)
idf = np.array(np.log((1.0*len(train_image_paths)+1) / (1.0*nbr_occurences+1)), 'float32')
```

```python
#Scaling the words
#Standardize features by removing the mean and scaling to unit variance (normalization)
from sklearn.preprocessing import StandardScaler
stdSlr = StandardScaler().fit(im_features)
im_features = stdSlr.transform(im_features)
```

```python
#Train SVM
from sklearn.svm import LinearSVC
clf = LinearSVC(max_iter=50000)
clf.fit(im_features, np.array(train_image_classes))
```

```python
np.unique(im_features)
```

```python
from sklearn.ensemble import AdaBoostRegressor
regr = AdaBoostRegressor(random_state=0, n_estimators=200)
regr.fit(im_features, np.array(train_image_classes))
```

```python
#Save the SVM
#joblib dumps python object into one file
import joblib
joblib.dump((clf, training_names, stdSlr, k, voc), 'bovw.pkl', compress=3)
#joblib.dump((regr, training_names, stdSlr, k, voc), 'bovw_ada.pkl', compress=3)
```
